````python
# Calculate the coefficient of best fit line or curve
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
y = [88,12,23,3,6,8,9,12,27,36,34,77,60,35,36,37,18,49,50,100,10,32,49,10,10,22]

print(np.polyfit(x,y,1))
````

````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    # ANSWER
    '''
    x_sum = sum(X)
    y_sum = sum(Y)
    x2_sum = sum([n**2 for n in X])
    xy_sum = sum(x * y for x, y in zip(X, Y))
    
     


    m = (5*xy_sum-(x_sum*y_sum))/(5*x2_sum - (x_sum**2))
    b = (y_sum - m*x_sum)/5
    
    return m, b

# solution
m, b = best_fit(X, Y)

print(m,b)
# #best fit line:
best_fit_line= b + m * np.array(X)

# # plot points and fit line
# # ANSWER


# Create a scatter plot for the data points
plt.scatter(X, Y, label='Data Points', c="red")

# # Plot the best-fit line
plt.plot(X, best_fit_line, color='red', label='Best Fit Line')

# Add labels and legend
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(visible=True)
plt.title('Scatter Plot with Best Fit Line')
plt.legend()

 # Display the plot
plt.show()
````

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m, h
# ANSWER
#m, se = ???, ???
#h = ???
#return m, m-h, m+h

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m, se = np.mean(a), scipy.stats.sem(a)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h
data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
# data = [170,167,	174,	179,	179,	187,	179,	183,	179 ,156,	163	,156	,187,	156	,
#         167	,156	,174	,170 ,183	,179	,174	,179	,170,	159	,187	]
print(mean_confidence_interval(data))

results = stats.ttest_1samp(data,170,alternative='greater')
````
