```python
# Create a scatter plot with regression line
sns.regplot(x="total_bill", y="tip", data=tips)
```

```python
# 'catplot()': Figure-level interface for drawing categorical plots onto a FacetGrid.
sns.catplot('Sex', data=Titanic, kind='count')
```

```python
weather.plot(y=['Tmax','Tmin'], x='Month')
```

```python
# Assuming 'weather' is a DataFrame with columns: 'Month', 'Tmax', and 'Tmin'
sns.lineplot(data=weather, x='Month', y='Tmax', label='Max Temperature')
sns.lineplot(data=weather, x='Month', y='Tmin', label='Min Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Max and Min Temperatures by Month')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_iris 
dataset=load_iris()
data=pd.DataFrame(dataset["data"],columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
data["Species"]=dataset["target"]
data["Species"]=data["Species"].apply(lambda x: dataset["target_names"][x])
print(data.head())
print(data.describe())
```

```python
sns.pairplot(data, hue="Species")
```
